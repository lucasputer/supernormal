```python
import math
import itertools
```

```python
def get_occurrences(amounts,total_words):
    #returns frequency of occurencies for each value in the observed range
    occs = {}
    observed_words = 0    
    for k,v in amounts.items():
        observed_words += 1
        if v not in occs:
            occs[v] = 0
        occs[v] = occs[v] + 1
    occs[0] = int(total_words - observed_words)
    max_o = 0
    for k,v in occs.items():
#         occs[k] = v/2**M
        occs[k] = v
        if k > max_o:
            max_o = k
    occs_array = [occs[i] if i in occs  else 0 for i in range(total_words+1)]
    return occs_array

def print_distributions(M):
    N = 2**M + M -1
    words = itertools.product(['0', '1'], repeat=N)
    distributions = {}
    finished = False
    total_words = 2**M
    while not finished:
        try:
            repetitions = {}
            w = ''.join(next(words))
            a = '1'
            for j in range(N - 1):
                a = a + '0'
            if w == a:
                finished = True
                print('f')
                break;
            for i in range(total_words):
                sub_word = w[i:i+M]
                if sub_word not in repetitions:
                    repetitions[sub_word] = 0
                repetitions[sub_word] = repetitions[sub_word] + 1
            occs = get_occurrences(repetitions,total_words)
            if occs[0] == 3 and occs[1] == 3:
                print(w)
            occs = str(occs)
            if occs not in distributions:
                distributions[occs] = 0
            distributions[occs] = distributions[occs] + 2
        except StopIteration:
            finished = True
    sorted_distributions = sorted(distributions.items(),key=lambda kv: kv[1])
    print('M = ' + str(M))
    for s in sorted_distributions:
        print(s[0] + ' : ' + str(s[1]))
    return sorted_distributions
```

```python
s = []
for i in range(5):
    s.append(print_distributions(i))
```

```python
def poisson_puntual(M,k):
    return 2**M*math.exp(-1)/math.factorial(k)
```

```python
for i in range(2*4):
    print(poisson_puntual(5,i))
```
